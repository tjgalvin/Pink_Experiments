# Figure showing source heatmap

Make a figure showing the heat map of a source for a trained SOM

```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import pandas as pd
import numpy as np
import pink_utils as pu
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
df = pd.read_csv('../Experiment/Experiment_F3W2_95_5/Test_Images/RGZ_Test_Images_Dataframe.csv')
```

```python
hm = pu.heatmap('../Experiment/Experiment_F3W2_95_5/F3W2_95_5_Testing_Images_EDMatrix.bin')
images = pu.image_binary('../Experiment/Experiment_F3W2_95_5/F3W2_95_5_Testing_Images.bin')
```

```python
index = 5454
src_map = hm.ed(index=index)
src_img = images.get_image(index)
src_img_wise = images.get_image(index, channel=1)
# src_img = binary.get_image(index=index, channel=0)
# src_map = binary.src_heatmap[SOM_path][index]
```

```python
color_map='bwr'
# title = binary.sources[index].rgz_path.split('/')[-1].replace('FIRST','').split('_')[0]
title = df.loc[index]['rgz_name']

fig, ax = plt.subplots(2,1, figsize=(5,8))

im = ax[0].imshow(src_img, cmap=plt.get_cmap(color_map))
# ax[0].set(title=f'$\\texttt{{{binary.channels[0]} {title}}}$')
ax[0].get_xaxis().set_ticks([])
ax[0].get_yaxis().set_ticks([])

divider = make_axes_locatable(ax[0])
cax0 = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im, cax=cax0, label='Intensity')

im = ax[1].imshow(src_map)
ax[1].get_xaxis().set_ticks([])
ax[1].get_yaxis().set_ticks([])

divider = make_axes_locatable(ax[1])
cax1 = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im, cax=cax1, label='Euclidean Distance')

pos = np.where(src_map==src_map.min())

ax[1].plot(*pos[::-1], 'ro')

fig.savefig('Images/Source_Heatmap.pdf')
```

```python
def plt_column(ax, src_img, src_img_wise, ed, count, index):
    title = df.loc[index]['rgz_name'].replace('RGZ','RGZ\,')
    label = df.loc[index]['label'].replace('_','\_')
    rcl = df.loc[index]['consensus.radio_level']
    
    title = f'$\\texttt{{{title}}}$\nLabel: {label} - RCL: {rcl:.2f}'
    im = ax[0].imshow(src_img, cmap=plt.get_cmap(color_map))
    ax[0].set(title=f'{title}')
    ax[0].get_xaxis().set_ticks([])
    ax[0].get_yaxis().set_ticks([])

    bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)
    ax[0].annotate(f'(a{count})', xy=(0.8,0.05), xycoords='axes fraction', bbox=bbox_props)

    divider = make_axes_locatable(ax[0])
    cax0 = divider.append_axes('right', size='5%', pad=0.05)
    fig.colorbar(im, cax=cax0, label='Intensity')

    im = ax[1].imshow(src_img_wise, cmap=plt.get_cmap(color_map))
    ax[1].get_xaxis().set_ticks([])
    ax[1].get_yaxis().set_ticks([])

    bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)
    ax[1].annotate(f'(b{count})', xy=(0.8,0.05), xycoords='axes fraction', bbox=bbox_props)

    divider = make_axes_locatable(ax[1])
    cax0 = divider.append_axes('right', size='5%', pad=0.05)
    fig.colorbar(im, cax=cax0, label='Intensity')

    im = ax[2].imshow(src_map)
    ax[2].get_xaxis().set_ticks([])
    ax[2].get_yaxis().set_ticks([])

    divider = make_axes_locatable(ax[2])
    cax1 = divider.append_axes('right', size='5%', pad=0.05)
    fig.colorbar(im, cax=cax1, label='Euclidean Distance')

    pos = np.where(src_map==src_map.min())
    ax[2].plot(*pos[::-1], 'ro')
    ax[2].annotate(f'(c{count})', xy=(0.8,0.05), xycoords='axes fraction', bbox=bbox_props)


    l = 1/src_map**10.
    l = l / l.sum()

    im = ax[3].imshow(l)
    ax[3].get_xaxis().set_ticks([])
    ax[3].get_yaxis().set_ticks([])

    divider = make_axes_locatable(ax[3])
    cax3 = divider.append_axes('right', size='5%', pad=0.05)
    mode = '%.2f' if count == 1 else '%.1f'
    fig.colorbar(im, cax=cax3, label='Likelihood', format=mode)

    pos = np.where(src_map==src_map.min())
    ax[3].plot(*pos[::-1], 'ro')
    ax[3].annotate(f'(d{count})', xy=(0.8,0.05), xycoords='axes fraction', bbox=bbox_props)

```

```python
fig, ax = plt.subplots(4,3, figsize=(2.3*3,6.5))

for c, index in enumerate([54,5455,7005]):
    src_map = hm.ed(index=index)
    src_img = images.get_image(index)
    src_img_wise = images.get_image(index, channel=1)

    plt_column(ax[:,c], np.sqrt(src_img), src_img_wise, src_map, c+1, index)

fig.subplots_adjust(hspace=0.003, wspace=0.4, bottom=0.01)
# fig.tight_layout()
# fig.show()
fig.savefig('Images/Figure_First_Wise_Heatmap_Likelihood_Columns.pdf')
```
