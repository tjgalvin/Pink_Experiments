```python
%matplotlib inline
# import matplotlib
# matplotlib.rcParams['text.usetex'] = True

import numpy as np
import pink_utils as pu
from collections import defaultdict
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
base= '../Experiment/Experiment_F3W2_95_5/Results/F3W2_95_5/F3W2S_95_5'

new = None
first = []
wise = []
stages = []
for c in [0]:
    count = 0
    for s in [1,2,3,4,5,6]:
        for i in range(20):
            som = pu.som(f'{base}_{s}_{i}.bin')
            print(som.path)
            im = som.get_som(channel=c)
            count += 1
            if c == 0:
                first.append(im)
            else:
                wise.append(im)
            if i == 0 and c == 0:
                stages.append(count)
```

```python
first = np.array(first)
wise = np.array(wise)
```

```python
first.shape
```

```python
import scipy.interpolate as interp
import matplotlib.animation as animation
from tqdm import tqdm
```

```python
blah = interp.interp1d(np.arange(120), first, axis=0)

fig, ax  = plt.subplots(1,1, figsize=(10,10))

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ims = []
# for i in tqdm(np.arange(120)):
for i in tqdm(np.linspace(0,119,400)):
    im = ax.imshow(np.sqrt(blah(i)), animated=True, cmap='bwr')
    ims.append([im])

fig.tight_layout()
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)

ani.save('dynamic_images.mp4')
```

```python
blah = interp.interp1d(np.arange(120), wise, axis=0)

fig, ax  = plt.subplots(1,1, figsize=(10,10))

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ims = []
# for i in tqdm(np.arange(120)):
for i in tqdm(np.linspace(0,119,400)):
    im = ax.imshow(np.sqrt(blah(i)), animated=True, cmap='bwr')
    ims.append([im])

fig.tight_layout()
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)

ani.save('wise_dynamic_images.mp4')
```

# Video for a specific neuron

```python
base= '../Experiment/Experiment_F3W2_95_5/Results/F3W2_95_5/F3W2S_95_5'

new = None
first = []
wise = []
stages = []
for c in [0]:
    count = 0
    for s in [1,2,3,4,5,6]:
        for i in range(20):
            som = pu.som(f'{base}_{s}_{i}.bin')
            print(som.path)
            im = som.get_neuron(channel=c, x=0, y=0)
            count += 1
            if c == 0:
                first.append(im)
            else:
                wise.append(im)
            if i == 0 and c == 0:
                stages.append(count)
```

```python
blah = interp.interp1d(np.arange(120), first, axis=0)

fig, ax  = plt.subplots(1,1, figsize=(10,10))

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ims = []
# for i in tqdm(np.arange(120)):
for i in tqdm(np.linspace(0,119,400)):
    ax.set(title=i)
    im = ax.imshow(np.sqrt(blah(i)), animated=True, cmap='bwr')
    ims.append([im])

fig.tight_layout()
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)

ani.save('dynamic_neuron_images.mp4')
```
