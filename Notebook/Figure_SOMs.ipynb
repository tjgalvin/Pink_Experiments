```python
chan_1 = som.get_som(channel=0)
dets = som.file_head

letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
x_locs = np.arange(dets[1])
x_locs *= dets[4]
x_locs += dets[4]//2

fig, ax = plt.subplots(1,1, figsize=(10,8.))
chan_1 = np.sqrt(chan_1)

im = ax.imshow(chan_1, cmap='bwr')# cmap='Greys')#, norm=mpl.colors.SymLogNorm(0.03))
# pu.no_ticks(ax)
ax.set(title='FIRST Channel')
ax.set(xticks=x_locs, xticklabels=np.arange(dets[1]))
ax.set(yticks=x_locs, yticklabels=[letters[i] for i in np.arange(dets[1])])

ax.set_xticks(np.arange(dets[1]+1) * dets[4], minor=True)
ax.set_yticks(np.arange(dets[1]+1) * dets[4], minor=True)

ax.grid(which='minor', color='black', linestyle='-', linewidth=2)

fig.colorbar(im, label='Intensity')
fig.tight_layout()
# fig.show()
fig.savefig('Images/SOM_First_Chan.pdf')
```
